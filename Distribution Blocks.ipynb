We use distribution plots to visulize the distribution of quantitative data.



```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

Displot/Histplot

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

# Joint plot

```python
sns.jointplot(x='total_bill', y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip',data=df,kind='reg')
```

# Pairplot

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```

# rugplot

```python
sns.rugplot(df)
```
